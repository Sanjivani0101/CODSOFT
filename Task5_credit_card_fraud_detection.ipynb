```python
# get the fraud and normal dataset
normal=credit_card_data[credit_card_data['Class']==0]
fraud=credit_card_data[credit_card_data['Class']==1]
```

Under Sampling

```python
# Build a sample dataset containing similar distribution of normal transactions and fraud transactions
normal_sample=normal.sample(n=492)
```

```python
# concatenating two DataFrames
new_dataset=pd.concat([normal_sample,fraud],axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

Splitting the data into featues and targets

```python
x=new_dataset.drop(columns='Class',axis=1)
y=new_dataset['Class'].astype(int)
```

```python
print(x)
```

```python
print(y)
```

To split the data into training data & testing data

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,stratify=y,random_state=2)
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

Model Training

```python
model=LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
pred=model.predict(x_test)
print('Accuracy score is :',accuracy_score(y_test,pred))
```

```python
# classification report
print(classification_report(y_test,pred))
```
